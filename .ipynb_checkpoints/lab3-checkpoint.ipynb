```python
def task_11():
    df = pd.read_csv("NationalNames.csv")
    year_unique_names = df.groupby("Year")["Name"].nunique()
    best_year = year_unique_names.sort_values(ascending=False).index[0]
    print(best_year)

task_11()
```

```python
def task_12():
    df = pd.read_csv("NationalNames.csv")
    year_unique_names = df.groupby("Year")["Name"].nunique()
    best_year = year_unique_names.sort_values(ascending=False).index[0]
    best_name = df[df["Year"] == best_year]["Name"].value_counts().index[0]
    print(best_name)

task_12()
```

```python
def task_13():
    df = pd.read_csv("NationalNames.csv")
    df = df[df["Gender"] == "F"]
    jacob_year = df[df["Name"] == "Jacob"]["Year"].value_counts().sort_values(ascending=False).index[0]
    print(jacob_year)

task_13()
```

```python
def task_14():
    df = pd.read_csv("NationalNames.csv")
    df_unisex = df[df["Gender"] == "F"].merge(df[df["Gender"] == "M"], how="inner", on=["Name", "Year"])
    year_unisex_names = df_unisex.groupby("Year")["Name"].nunique()
    best_year = year_unisex_names.sort_values(ascending=False).index[0]
    print(best_year)

task_14()
```

```python
def task_17():
    df = pd.read_csv("NationalNames.csv")
    year_gender_count = df.groupby(["Year", "Gender"])["Count"].sum()
    print(year_gender_count)

task_17()
```

```python
def task_18():
    df = pd.read_csv("NationalNames.csv")
    year_gender_count = df.groupby(["Year", "Gender"])["Count"].sum()

    if "F" in year_gender_count.index:
        girls_more = len(year_gender_count[(year_gender_count["F"] > year_gender_count["M"])])
    else:
        girls_more = 0 
    print(girls_more)

task_18()
```

```python
def task_23():
    df = pd.read_csv("NationalNames.csv")
    df_f = df[df["Gender"] == "F"]
    df_m = df[df["Gender"] == "M"]
    unisex_names = set(df_f["Name"].unique()) & set(df_m["Name"].unique())
    print(unisex_names)

task_23()
```

```python
def task_24():
    df = pd.read_csv("NationalNames.csv")
    name_counts = df["Name"].value_counts()
    least_popular = name_counts.sort_values(ascending=False).index[-1]
    count_least_popular = name_counts.iloc[-1]
    print(f"Name: {least_popular}, Count: {count_least_popular}")

task_24()
```

```python
import matplotlib.pyplot as plt

def task_26():
  df = pd.read_csv("NationalNames.csv")
  john_counts = df[(df["Name"] == "John") & (df["Gender"] == "F")]["Year"].value_counts()
  mary_counts = df[(df["Name"] == "Mary") & (df["Gender"] == "M")]["Year"].value_counts()
  john_counts.plot(kind="line", label="John (Female)")
  mary_counts.plot(kind="line", label="Mary (Male)")
  plt.xlabel("Year")
  plt.ylabel("Count")
  plt.title("Distribution of John (Female) and Mary (Male) Names")
  plt.legend()

  # Display the plot
  plt.show()

task_26()
```
