```python
import pandas as pd
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head(4))
```

```python
def count_gender():
    gender_counts = df['Gender'].value_counts()
    return gender_counts['male'], gender_counts['female']

male_count, female_count = count_gender()

# 1. Скільки студентів і студенток у наборі даних?
print("Number of male students:", male_count)
print("Number of female students:", female_count)
```

```python
# 2. Який відсоток студентів чоловічої статі в наборі даних?

def percentage_male():
    male_count, female_count = count_gender()
    total_students = male_count + female_count
    return (male_count / total_students) * 100

print("Percentage of male students:", percentage_male(), "%")
```

```python
# 3. Який відсоток студенток у наборі даних?
def percentage_female():
    male_count, female_count = count_gender()
    total_students = male_count + female_count
    return (female_count / total_students) * 100

print("Percentage of female students:", percentage_female(), "%")
```

```python
def calculate_average_scores():
    avg_math_scores = df.groupby('EthnicGroup')['MathScore'].mean()
    avg_reading_scores = df.groupby('EthnicGroup')['ReadingScore'].mean()
    avg_writing_scores = df.groupby('EthnicGroup')['WritingScore'].mean()
    return avg_math_scores, avg_reading_scores, avg_writing_scores
```

```python
def highest_math_score_ethnic_group():
    avg_math_scores, _, _ = calculate_average_scores()
    return avg_math_scores.idxmax()
def highest_reading_score_ethnic_group():
    _, avg_reading_scores, _ = calculate_average_scores()
    return avg_reading_scores.idxmax()
def highest_writing_score_ethnic_group():
    _, _, avg_writing_scores = calculate_average_scores()
    return avg_writing_scores.idxmax()

# 4. Яка етнічна група має найвищий середній бал з математики?
highest_math_group = highest_math_score_ethnic_group()
print("Ethnic group with the highest average math score:", highest_math_group)

# 5. Яка етнічна група має найвищий середній бал читання?
highest_reading_group = highest_reading_score_ethnic_group()
print("Ethnic group with the highest average reading score:", highest_reading_group)

# 6. Яка етнічна група має найвищий середній бал з письма?
highest_writing_group = highest_writing_score_ethnic_group()
print("Ethnic group with the highest average writing score:", highest_writing_group)
```

```python
def average_scores_test_prep():
    avg_math_score = df[df['TestPrep'] == 'completed']['MathScore'].mean()
    avg_reading_score = df[df['TestPrep'] == 'completed']['ReadingScore'].mean()
    avg_writing_score = df[df['TestPrep'] == 'completed']['WritingScore'].mean()
    return avg_math_score, avg_reading_score, avg_writing_score

def count_parents_marital_status():
    married_parents_count = df[df['ParentMaritalStatus'] == 'married'].shape[0]
    single_parents_count = df[df['ParentMaritalStatus'] != 'married'].shape[0]
    return married_parents_count, single_parents_count
```

```python
# 7. Який середній бал з математики для студентів, які закінчили курс підготовки до тесту?
avg_math_prep = df[df['TestPrep'] == 'completed']['MathScore'].mean()
print("Average math score for students who completed the test preparation course:", avg_math_prep)

# 8. Який середній бал з читання для студентів, які пройшли курс підготовки до тесту?
avg_reading_prep = df[df['TestPrep'] == 'completed']['ReadingScore'].mean()
print("Average reading score for students who completed the test preparation course:", avg_reading_prep)

# 9. Який середній бал для студентів, які закінчили курс підготовки до тесту?
avg_writing_prep = df[df['TestPrep'] == 'completed']['WritingScore'].mean()
print("Average writing score for students who completed the test preparation course:", avg_writing_prep)

married_count, single_count = count_parents_marital_status()

# 10. Скільки студентів мають одружених батьків?     
print("Number of students with married parents:", married_count)

# 11. Скільки студентів мають батьків-одинаків?
print("Number of students with single parents or other marital status:", single_count)
```

```python
def average_scores_sports():
    avg_math_score_sports = df[df['PracticeSport'] == 'regularly']['MathScore'].mean()
    avg_reading_score_sports = df[df['PracticeSport'] == 'regularly']['ReadingScore'].mean()
    return avg_math_score_sports, avg_reading_score_sports

def average_scores_no_sports():
    avg_math_score_no_sports = df[df['PracticeSport'] == 'never']['MathScore'].mean()
    avg_reading_score_no_sports = df[df['PracticeSport'] == 'never']['ReadingScore'].mean()
    return avg_math_score_no_sports, avg_reading_score_no_sports
```

```python
avg_math_sports, _ = average_scores_sports()
avg_math_no_sports, _ = average_scores_no_sports()

# 12. Чи студенти, які регулярно займаються спортом, мають вищі середні результати з математики, ніж ті, хто ніколи не       займається спортом?
print("Average math score for students who regularly practice sports:", avg_math_sports)
print("Average math score for students who never practice sports:", avg_math_no_sports)
print("Students who regularly practice sports have higher average math scores.") if avg_math_sports > avg_math_no_sports else print("Students who never practice sports have higher average math scores.")
```

```python
_, avg_reading_sports = average_scores_sports()
_, avg_reading_no_sports = average_scores_no_sports()

# 13. Чи студенти, які регулярно займаються спортом, мають вищі середні результати читання, ніж ті, хто ніколи не займається спортом?
print("Average reading score for students who regularly practice sports:", avg_reading_sports)
print("Average reading score for students who never practice sports:", avg_reading_no_sports)
print("Students who regularly practice sports have higher average reading scores.") if avg_reading_sports > avg_reading_no_sports else print("Students who never practice sports have higher average reading scores.")
```

```python
def count_school_bus():
    return df[df['TransportMeans'] == 'school_bus'].shape[0]

def count_private_transport():
    return df[df['TransportMeans'] == 'private'].shape[0]
```

```python
# 14. Скільки учнів користується шкільним автобусом, щоб дістатися до школи?
school_bus_count = count_school_bus()
print("Number of students using the school bus to get to school:", school_bus_count)

# 15. Скільки учнів користується приватним транспортом, щоб дістатися до школи?
private_transport_count = count_private_transport()
print("Number of students using private transportation to get to school:", private_transport_count)
```

### Передбачте, чи завершив студент курс підготовки до іспиту, на основі його демографічних та академічних даних.
Вхідні дані:
Gender, EthnicGroup, ParentEduc, LunchType, ParentMaritalStatus, PracticeSport, IsFirstChild, NrSiblings,       TransportMeans, WklyStudyHours, MathScore, ReadingScore, WritingScore
Результат: TestPrep

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier

# Convert categorical variables to numeric
le = LabelEncoder()
df = df.apply(le.fit_transform)

# Define the features and the target
features = df[['Gender', 'EthnicGroup', 'ParentEduc', 'LunchType', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild', 'NrSiblings', 'TransportMeans', 'WklyStudyHours', 'MathScore', 'ReadingScore', 'WritingScore']]
target = df['TestPrep']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

# Create a base model
rf = RandomForestClassifier(random_state=42)

# Instantiate the grid search model
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_

print("Best parameters: ", best_params)

# Train the model using the best parameters
model = RandomForestClassifier(**best_params)
model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```
